```python
# EDA and Cleaning

# This notebook imports the data, runs initial exploration, and prepares for cleaning.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display settings for EDA
pd.options.display.max_columns = None
pd.options.display.max_rows = 100
sns.set(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
data_path = 'complaints-2026-06-30_00_09.csv'

# Load the dataset
df = pd.read_csv(data_path)

print('Dataset loaded successfully')
print('Shape:', df.shape)

df.head()
```

```python
# Basic EDA
print('Columns:', len(df.columns))
print(df.columns.tolist())

print('\nDataframe info:')
df.info()

print('\nSummary statistics:')
display(df.describe(include='all').T)

missing = df.isna().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print('\nMissing values by column:')
print(missing if not missing.empty else 'No missing values detected')

duplicates = df.duplicated().sum()
print(f'\nDuplicate rows: {duplicates}')

cat_cols = df.select_dtypes(include='object').columns.tolist()
num_cols = df.select_dtypes(include='number').columns.tolist()
print(f'\nCategorical columns ({len(cat_cols)}):', cat_cols)
print(f'Numeric columns ({len(num_cols)}):', num_cols)

if len(cat_cols) > 0:
    print('\nTop categories for first object columns:')
    for col in cat_cols[:3]:
        print(f'\n{col} value counts:')
        display(df[col].value_counts(dropna=False).head(10))

if len(num_cols) > 0:
    print('\nNumeric summary plots:')
    df[num_cols].hist(bins=20, layout=(max(1, len(num_cols)//3), 3), figsize=(15, 10), edgecolor='black')
    plt.tight_layout()
    plt.show()

    corr = df[num_cols].corr()
    if corr.shape[0] > 1:
        plt.figure(figsize=(10, 8))
        sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
        plt.title('Numeric Feature Correlation')
        plt.show()
```

```python
df['Date received'] = pd.to_datetime(df['Date received'])
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])

# Derive a useful feature: how long internally before CFPB routed it to company
df['days_to_route'] = (df['Date sent to company'] - df['Date received']).dt.days
```

```python
df['Tags'] = df['Tags'].fillna('None')
df['Company public response'] = df['Company public response'].fillna('No public response')
df['Sub-issue'] = df['Sub-issue'].fillna('Not specified')
df['State'] = df['State'].fillna('Unknown')
df['ZIP code'] = df['ZIP code'].fillna('Unknown')

# Sanity check - should be zero missing now except narrative isn't missing at all
print(df.isnull().sum())
```

```python
df['narrative_word_count'] = df['Consumer complaint narrative'].str.split().str.len()

print(df['narrative_word_count'].describe())

# Visualize distribution to pick a sensible cutoff
plt.figure(figsize=(8,4))
sns.histplot(df['narrative_word_count'], bins=60)
plt.xlim(0, 600)
plt.axvline(20, color='red', linestyle='--', label='cutoff = 20 words')
plt.legend()
plt.title('Narrative word count distribution')
plt.show()
```

```python
print("Rows below 20 words:", (df['narrative_word_count'] < 20).sum())
print("Rows below 10 words:", (df['narrative_word_count'] < 10).sum())

df_clean = df[df['narrative_word_count'] >= 20].copy()
print("Remaining rows:", df_clean.shape[0])
```

```python
# Drop negligible in-progress rows (not yet resolved, not useful for outcome analysis)
df_clean = df_clean[df_clean['Company response to consumer'] != 'In progress'].copy()

# Binary outcome for later validation (NOT used in clustering)
df_clean['got_relief'] = df_clean['Company response to consumer'].isin(
    ['Closed with monetary relief', 'Closed with non-monetary relief']
).astype(int)

print(df_clean['got_relief'].value_counts(normalize=True))

# Scope to 3 related, high-volume products
keep_products = ['Debt collection', 'Checking or savings account', 'Credit card']
df_scoped = df_clean[df_clean['Product'].isin(keep_products)].copy()

print(df_scoped.shape)
print(df_scoped['Product'].value_counts())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Complaint volume over time
df_scoped.set_index('Date received').resample('W').size().plot(ax=axes[0,0])
axes[0,0].set_title('Weekly complaint volume')
axes[0,0].set_ylabel('Count')

# 2. Top 10 companies by complaint volume
df_scoped['Company'].value_counts().head(10).plot(kind='barh', ax=axes[0,1])
axes[0,1].set_title('Top 10 companies by complaint volume')
axes[0,1].invert_yaxis()

# 3. got_relief rate by product
df_scoped.groupby('Product')['got_relief'].mean().sort_values().plot(kind='barh', ax=axes[1,0])
axes[1,0].set_title('Relief rate by product')
axes[1,0].axvline(df_scoped['got_relief'].mean(), color='red', linestyle='--')

# 4. days_to_route distribution
sns.histplot(df_scoped['days_to_route'], bins=30, ax=axes[1,1])
axes[1,1].set_title('Days to route distribution')

plt.tight_layout()
plt.show()
```

```python
for product in keep_products:
    print(f"\n=== {product} ===")
    print(df_scoped[df_scoped['Product']==product]['Issue'].value_counts().head(5))
```

```python
print(df_scoped['Date received'].max())

# Trim last ~45 days to avoid resolution-censoring bias
cutoff_date = df_scoped['Date received'].max() - pd.Timedelta(days=45)
df_final = df_scoped[df_scoped['Date received'] <= cutoff_date].copy()

print(df_final.shape)
print(df_final['got_relief'].mean())
print(df_final['Date received'].min(), '-', df_final['Date received'].max())
```

```python
print(df_final.groupby('Product')['got_relief'].mean().sort_values())
print()
print(df_final['Product'].value_counts())
```

```python
df_final.to_csv('cfpb_cleaned.csv', index=False)
print("Saved:", df_final.shape)
```
